[Crowdsourced EV Charging Stations](https://catalog.data.gov/dataset/potential-ev-charging-stations)

kmeans, cluster by geo for x number of possible stations and observe distribution?

```
import pandas as pd
```

```
data = pd.read_csv('/EV_Charging_Stations_Crowdsourced_Data_minnesota_p1.csv')
data.head(3)
```

```
#locations #list to dict to df
l = pd.DataFrame({'x': data['Latitude'], 'y': data['Longitude']})
l.head()
```

```
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters = 20).fit(l)
centroids = kmeans.cluster_centers_
print(centroids)

plt.scatter(l['x'], l['y'], c = kmeans.labels_.astype(float), s = 50, alpha = 0.5)
plt.scatter(centroids[:, 0], centroids[:, 1], c = 'red', s = 50)
plt.show()
```
